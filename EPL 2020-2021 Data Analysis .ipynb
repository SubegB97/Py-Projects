```python
# Import Libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
# Load Dataset 
# Created a variable for dataframe & read in file 


epl_df = pd.read_csv('C:/Users/Subeg/Documents/EPL_20_21.csv')

epl_df.head()
```

```python
# Cheking the info of the dataset 
epl_df.info()
```

```python
# Checking description of the dataset 
epl_df.describe()
```

```python
# Finidng Null Values 

epl_df.isna().sum()
```

```python
# Adding 2 more columns (minutes per match & goals per match)

epl_df['MinutesPerMatch'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals']/epl_df['Matches']).astype(float)
epl_df.head()
```

```python
# Computing overall total goals scored 

total_goals = epl_df['Goals'].sum()
total_goals
```

```python
total_penalty = epl_df['Penalty_Goals'].sum()
total_penalty 
```

```python
total_penalties_attempted = epl_df['Penalty_Attempted'].sum()
total_penalties_attempted
```

```python
# Pie Chart showing penalties scored vs missed 
plt.figure(figsize=(14,5))
pl_not_scored = epl_df['Penalty_Attempted'].sum() - total_penalty
data = [pl_not_scored,total_penalty]
labels = ['Penalties missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
#Unique Positions Played
epl_df['Position'].unique ()
```

```python
# Total players that play forward
epl_df[epl_df['Position'] == 'FW']
```

```python
#Players from different nations
np.size((epl_df['Nationality'].unique()))
```

```python
# Most players from which countries 
nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12,6), color = sns.color_palette("magma"))
```

```python
#Clubs with the most players in their squad 

epl_df['Club'].value_counts().nlargest(5).plot(kind = 'bar', color=sns.color_palette("viridis"))
```

```python
#Clubs with the least players in their squad 

epl_df['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color=sns.color_palette("viridis"))
```

```python
#Players based on age group 
#<20, 20-25, 25-30, 30>

under20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[epl_df['Age'] > 20 & (epl_df['Age'] <= 25)]
age25_30 = epl_df[epl_df['Age'] > 25 & (epl_df['Age'] <= 30)]
over30 = epl_df[epl_df['Age'] > 30]

x = np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),over30['Name'].count()])
mylabels = ["<20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('Total Players with Age', fontsize=20)
plt.pie(x, labels = mylabels, autopct="%.1f%%")
plt.show()
```

```python
#Total Players under 20 in each club 

players_under_20 = epl_df[epl_df ['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette("cubehelix"))
```

```python
#Average age of players in each club (Box Plot)

plt.figure(figsize=(12,6))
sns.boxplot(x = 'Club', y='Age', data = epl_df)
plt.xticks(rotation = 90)
```

```python
goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club',y='Goals',data=goals_by_clubs.sort_values(by="Goals"), palette='rocket')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize = 20)
```
