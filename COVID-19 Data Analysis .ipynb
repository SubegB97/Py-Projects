```python
#Dropping columns from first dataset 
#inplace = true means it will return nothing and the dataframe is now updated 
#axis = 1 means along "columns"
covid_df.drop(["Sno","Time","ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df.head()
```

```python
#Changing format of date column using todatetime 
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
#Finding the number of active cases 
#Confirmed cases - deaths + cured 
#Added a new column

covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured']+covid_df['Deaths'])
covid_df.tail(5)
```

```python
#Pivot Table 
#Summing all confirmed deaths and cured cases for each states/union territoies 
# Creating varaible to piv

statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
```

```python
#Finding out the recovery rate 
#Total number of cured cases divded by total number of confirmed cases 
statewise["Recovery Rate"] = statewise ["Cured"] * 100/statewise["Confirmed"]
```

```python
#Finding out the mortality rate 
#Total number of deaths divded by total number of confirmed cases 
statewise["Mortality"] = statewise ["Deaths"] * 100/statewise["Confirmed"]
```

```python
#Sorting Values based on Confirmed cases, decending order 
#Using sort function by passing in Column and Sort order 
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
#Plot Pivot Table using visual 

statewise.style.background_gradient(cmap =  "cubehelix")
```

```python
#Top 10 Active Cases States 
#Using group by pandas function 

top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by = ['Active_Cases'], ascending = False).reset_index()
```

```python
#Passing in x axis and y axis
#Use bar plot from seaborn library 
# iloc is index location, we're getting the top 10 
#Setting figure variable size
top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_Cases', 'Date']].sort_values(by = ['Active_Cases'], ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with most active Cases in India", size = 25) 
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y="Active_Cases", x="State/UnionTerritory", linewidth = 2, edgecolor = 'black')

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#Top Ten States with Most Deaths 
top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
fig = plt.figure(figsize=(15,5))
plt.title("Top 10 States with Deaths in India", size = 25) 
ax = sns.barplot(data = top_10_deaths.iloc[:12], y="Deaths", x="State/UnionTerritory", linewidth = 2, edgecolor = 'black')

plt.xlabel("States")
plt.ylabel("Deaths")
plt.show()
```

```python
#Line plot to show growth/trend of confirmed cases in top 5 
#Growth Trend 

fig = plt.figure(figsize=(12,6))

ax = sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])],x = 'Date', y = 'Active_Cases', hue = 'State/UnionTerritory')

ax.set_title("Top 5 Affected States in India", size = 16)

```
